# Tuberculosis

```python
tb = pd.read_csv("tb.csv")
tb.head()
```

## Error
 Multiple variables stored in one column, for example, in the column new_sp_m514, contains
 more than one variable; interval of age, sex, etc, and this happens to more columns

```python
#Here I use the melt function so my columns gets into the values
pew_tidy_tb = pd.melt(tb,id_vars=['iso2','year'],var_name='Colum')
#Here I add the "sex" column
pew_tidy_tb['Sex'] = pew_tidy_tb.Colum.str.slice(7,8)
#"Here I add the "age" column
pew_tidy_tb['age'] = pew_tidy_tb.Colum.str.slice(8,)
#Here I replace the m and f for male and female 
pew_tidy_tb.Sex = pew_tidy_tb.Sex.astype(str).replace('m','Male')
pew_tidy_tb.Sex = pew_tidy_tb.Sex.replace('f','Female')
```

```python
#We tried to do this part with "iterrows" or "".loc", but this was the most efficient way, we sware
ages = pew_tidy_tb.age.unique().tolist()
for entry in ages:
    if len(entry) == 2:
        if '6' not in entry:
            pew_tidy_tb.age = pew_tidy_tb.age.str.replace(entry,entry[0]+'-'+entry[1:])
        else:
            pew_tidy_tb.age = pew_tidy_tb.age.str.replace('65','65+')
    if len(entry) == 3:
        pew_tidy_tb.age = pew_tidy_tb.age.str.replace(entry,entry[0]+'-'+entry[1:])
    if len(entry) == 4:
        pew_tidy_tb.age = pew_tidy_tb.age.str.replace(entry,entry[0:2]+'-'+entry[2:])
```

```python
#Here I delete the column named "Column", we already obtained all the information from it
del pew_tidy_tb['Colum']
```

```python
#Here we changed the names of the columns
names = {'iso2':'Country','value':'Cases','year':'Year'}
pew_tidy_tb = pew_tidy_tb.rename(columns=names)
```

# Weather

```python
#read csv
weather = pd.read_csv("weather.txt", sep = '\t')
weather.head()
```

## Error 
Variables are stored in both rows and columns, this can be easy to see in the column element because contains variables that are TMAX and TMIN, should be columns. Additionally, id, year and month are variable columns by itself

```python
#read csv
weather = pd.read_csv("weather.txt", sep = '\t')
"""
melting the csv to get the column element that contains TMAX and TMIN and each one its temperature, and another
columns that contains that day, the drop the row that have empty values
"""
weather = pd.melt(weather, id_vars = ["id","year","month","element"], var_name = "Day", value_name = "Temperature").dropna().reset_index(drop = True)
#make the column Day into a int, eliminating the d that has
weather.Day = weather.Day.str.replace("d","").astype(int)
"""
getting the tidy table applying a pivot table, where the index are id, year, month and day
the columns are TMAX and TMIN, and the values the temperature
"""
weather = pd.pivot_table(weather, index = ['id','year','month','Day'], columns = 'element', values = 'Temperature').reset_index(drop= False).reset_index(drop= True)
#making a new column that has the full date
weather['Date'] = weather.year.astype(str)+"-"+weather.month.astype(str)+"-"+weather.Day.astype(str)
#eliminating no more useful columns
weather = weather.drop(['year','month','Day'], axis = 1)
weather.head()
```
