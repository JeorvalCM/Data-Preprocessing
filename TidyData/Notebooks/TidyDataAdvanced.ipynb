by having this error I noticed the T, so I searched and I found out this


In meteorology, the word "trace" is used to describe a very small amount of precipitation that results in no measurable accumulation. In other words, a 'trace' is when you can observe that some amount of rain or snow fell, but it was not enough to be measured using a rain gauge, snow stick, or any other weather instrument.

Since trace precipitation falls as very light and brief sprinkles or flurries, you often won't know it unless you happen to be outdoors and see or feel it falling. 

* Trace amounts of precipitation are abbreviated by the capital letter "T", often placed in parenthesis (T).
* If you must convert a trace to a numerical amount, it would equal 0.00.



So, I need to change all the T for zeros

```python
#changing the T into 0
for x in weather.columns[3:]:
    weather[x].loc[weather[x] == 'T'] = 0
    weather[x] = weather[x].astype(float)
```

```python
for j in range(1, int(len(weather_pivot.columns.tolist()[2:]) / 22)):
    x = weather_pivot.columns.tolist()[0:2]
    y = weather_pivot.columns.tolist()[(j*22)+2:(j+1)*22 + 2]
    x.extend(y)
    #useful columns
    columnas = ['year','month']
    #eliminating multi index
    droplevelofday = weather_pivot[x].droplevel(0, axis = 1).copy()
    #adding the columns we will use for renaming
    columnas.extend(droplevelofday.columns.tolist()[2:])
    #renaming
    droplevelofday.columns = columnas
    #getting the day of the X by list operations
    droplevelofday['Day'] = (weather_pivot.columns.tolist()[(j*22)+2:(j+1)*22 + 2][0][0])[1:]
    #aggregating the day to the final dataset
    weather_final = weather_final.append(droplevelofday, sort = False)
```

```python
#getting the data
weather_final['date: Y,M,D'] = weather_final.year.astype(str)+'-'+weather_final.month.astype(str)+'-'+weather_final.Day.astype(str)
```

```python
#transforming the data into int
weather_final.Day = weather_final.Day.astype(int)
```

```python
#sorting values
weather_final = weather_final.sort_values(['year','month','Day'])
```

```python
#setting new index
weather_final.set_index('date: Y,M,D', inplace = True)
```
