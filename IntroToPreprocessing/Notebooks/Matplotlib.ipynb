# Plotting with Mathplotlib

```python
from matplotlib import pyplot as plt
import pandas as pd

# Global option for pyplot
plt.rcParams['figure.figsize'] = [12,6]
```

```python
#load the dataframe
df = pd.read_csv('Kumpula-June-2016-w-metadata.txt', skiprows=8)
```

```python
date = df['YEARMODA'].values
temp = df['TEMP'].values
temp_max = df['MAX'].values
temp_min = df['MIN'].values
```

## A simple plott

```python
plt.plot(date, temp)
```

## Basic plot formating

```python
plt.plot(date, temp, 'ro--')
plt.title('Kumpula temperatures in June 2016')
plt.xlabel('Date')
plt.ylabel('Temperature[°F]')
plt.text(20160604, 68, ' High temperature in early June')
```

```python
plt.plot(date, temp, 'ro--')
plt.title('Kumpula temperatures in June 2016')
plt.xlabel('Date')
plt.ylabel('Temperature[°F]')
plt.text(20160604, 68, ' High temperature in early June')
plt.axis([20160601, 20160615, 55, 70])
```

```python
plt.plot(date, temp, 'ro--')
plt.title('Kumpula temperatures in June 2016')
plt.xlabel('Date')
plt.ylabel('Temperature[°F]')
plt.text(20160604, 68, ' High temperature in early June')
plt.xlim([20160601, 20160615])
```

```python
plt.plot(date, temp, 'ro--')
plt.title('Kumpula temperatures in June 2016')
plt.xlabel('Date')
plt.ylabel('Temperature[°F]')
plt.text(20160604, 68, ' High temperature in early June')
plt.ylim([55, 70])
```
