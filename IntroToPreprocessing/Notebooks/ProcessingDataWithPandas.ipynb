# Processing data  with pandas

```python
# importing the modules needed
import pandas as pd
import numpy as np
```

## Loading data  

```python
fname = r'Kumpula-June-2016-w-metadata.txt'
dataframe = pd.read_csv(fname, sep=',', skiprows=8)
```

## Calculating
 - creating a new column called "DIFF" and populate it with the difference between the MAX and MIN
 - creating a new column called "DIFF_Min" and populate it with the difference between the TEMP and MIN
 - creating a new column called "TEMP_CELCIUS" and populate it with the conversion from "F to Â°C of the "TEMP" column

```python
# using comprenhension lists
"""
dataframe['DIFF'] = [dataframe["MAX"][x]-dataframe["MIN"][x] for x in range(0,30)]
dataframe["DIFF_Min"] = [dataframe["TEMP"][x]-dataframe["MIN"][x] for x in range(0,30)]
dataframe["TEMP_CELCIUS"] = [(dataframe["TEMP"][x]-32)*(5/9) for x in range(0,30)]
dataframe
"""
```

```python
# using Pandas
dataframe['DIFF'] = dataframe["MAX"]-dataframe["MIN"]
dataframe['DIFF_Min'] = dataframe["TEMP"]-dataframe["MIN"]
dataframe['TEMP_CELSIUS'] = (dataframe["TEMP"]-32)*(5/9)
dataframe
```

## Filtering and updating data

```python
#select rows with a temp_celsius value greater then 15 degrees
w_temp = dataframe.loc[dataframe['TEMP_CELSIUS'] > 15]
w_temp
```

```python
# select rows with a TEMP_CELSIUS is greter than 15 degrees and from the second half of the year
w_temp2 = dataframe.loc[(dataframe['TEMP_CELSIUS'] > 15) & (dataframe['YEARMODA'] >= 20160615) ]
w_temp2
```

```python
w_temp2 = w_temp2.reset_index(drop=True)
w_temp2
```

# sortign tha data

```python
# sort in asecnding order the dataframe according to a column, for example 'TEMP'
sorted_temp_df = dataframe.sort_values(by='TEMP')
sorted_temp_df
```

```python
# sorting the dataframe ip ascending order
sorted_temp2_df = dataframe.sort_values(by='TEMP',ascending = False)
sorted_temp2_df
```

# Rounding and finding unique values

```python
# create new column with the temperature in Celsius rounden to the nearest integer
dataframe['Celsius_rounded'] = dataframe['TEMP_CELSIUS'].round(0).astype(int)
dataframe
```

```python
# getting the unique values in the celsius_rounded column
unique = dataframe['Celsius_rounded'].unique()
unique
```

## Writing the data to a file

```python
# name of the output file
outfname = 'output1.csv'

#save dataframe to csv
dataframe.to_csv(outfname,sep=',')
```

```python
# name of the output file
outfname2 = "output2.csv"

#save the dataframe to csv, with custom formating for the float numbers and without index
dataframe.to_csv(outfname2, sep=',', index=False, float_format="%.2f")
```

```python
# saving the dataframe DIRECTLY to excel
# name of the output file
outfname3 = "output3.xlsx"

#initialize ExcelWriter
writer = pd.ExcelWriter(outfname3)

#save the dataframe to csv, with custom formating for the float numbers and without index
dataframe.to_excel(writer, sheet_name="Kumpula_temperatures", index=False, float_format="%.2f")
writer.save()
```
