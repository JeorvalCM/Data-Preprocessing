## Iterating rows and using custom functions

```python
def fahrToCelsius(temp_f):
    '''
    docstring
    '''
    
    temp_c = (temp_f-32)/1.8
    return temp_c
```

```python
# iterate over the rows
for idx, row in data.iterrows():
    print("Id: ",idx,"row: ",row)
    print(type(row))
    break
```

```python
#creating the column in the dataframe
col_name = 'Celsius'
data[col_name] = None

#iterating over the rows
for idx, row in data.iterrows():
    # make the conversion
    celsius = fahrToCelsius(row['TEMP'])
    
    #Update the temperature in the celsius columns
    data.loc[idx, col_name] = celsius
```

```python
data.head()
```

```python
# convert the speed from mph to m/s
# note: m/s = mph * 0.44704
data['SPEED'] = data['SPEED']*0.44704
data['GUST'] = data['GUST']*0.44704

data.head()
```

## String manipulation

```python
# converting time column into a string
data['TIME'] = data['TIME'].astype('str')
```

```python
# Time column with time up to hours
data['TIME_dh']  = data['TIME'].str.slice(start=0,stop=10)
data.head()
```

```python
# create a new colummn with the hour as a numeric column
data['TIME_h'] = data['TIME'].str.slice(start=8,stop=10).astype(int)
data.head()
```

## Aggregating data by grouping 

```python
# Group the data bases on the time up to hours
grouped = data.groupby('TIME_dh')
grouped
```

```python
# check an specific time
time1 = '2017080400'

#make the slection for the group
group1 = grouped.get_group(time1)
group1
```

```python
# specify the columns that will be averaged
mean_cols = ['DIR','SPEED','GUST','TEMP','Celsius','TIME_h']

# calculate the means
mean_values = group1[mean_cols].mean()
mean_values
```

```python
# create a new empty dataframe to stored the aggregated data
aggr_data = pd.DataFrame()

# append the values in our new dataframe
aggr_data = aggr_data.append(mean_values, ignore_index = True)
aggr_data
```

```python
# iterate over groups
for key, group in grouped:
    print('Key:\n',key)
    print('\nGroup: \n', group)
```

```python
# create an empty dataframe for the agregated values
aggr_data = pd.DataFrame()

# define the columns to be aggregated
mean_cols = ['DIR','SPEED','GUST','TEMP','Celsius','TIME_h']

#iterate over the groups
for key, group in grouped:
    # calculate the mean
    mean_values = group[mean_cols].mean()
    #add the key
    mean_values['TIME_dh'] = key
    
    # append the series to the dataframe
    aggr_data = aggr_data.append(mean_values, ignore_index = True)
    
aggr_data   
```

## Finding outliers from the data
We define and outlier if the value is 2 times the standard deviation higher than the average.(SPEED column)

```python
# calculate the standard deviation and average wind speed
std_wind = aggr_data['SPEED'].std()
avg_wind = aggr_data['SPEED'].mean()
print('The std. dev is: ', std_wind)
print('The average is: ', avg_wind)
```

```python
# calculate the upper threshold for an outlier
upper_treshold = avg_wind + (2*std_wind)
print("The upper treshold for an outlier is: ", upper_treshold)
```

Create a new column called 'Oulier' and fill it with True or False, accordingly

```python
aggr_data['Outlier'] = aggr_data['SPEED'] > upper_treshold
aggr_data
```

```python
# select rows that are outliers
storm = aggr_data.loc[aggr_data['Outlier'] == True]
storm
```
