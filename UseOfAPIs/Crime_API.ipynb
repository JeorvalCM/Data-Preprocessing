```python
import json
import pandas as pd
import requests
```

```python
def crimes_types(): 
    """
    docstring
    description: Function that gets all the crymes types and return a python dictioary
    """
    url = 'https://hoyodecrimen.com/api/v1/crimes'
    res = requests.get(url)
    results = json.loads(res.content)
    return results
```

```python
crimes = crimes_types()
```

```python
def all_cuadrants():
    """
    docstring
    description: Function that gets all the cuadrants from mexico city
    """
    url = 'https://hoyodecrimen.com/api/v1/cuadrantes'
    res = requests.get(url)
    results = json.loads(res.content)
    return results
```

```python
cuadrants = all_cuadrants()
cuadrants
```

```python
def crimes_loc(coordinates, distance):
    """
    Docstring
    Description: function that gets the crimes given coordinates
    Input: coordinates of the location in lat and lon, distance radius 
    output: return a python dictionaty with that data get
    """
    url = 'https://hoyodecrimen.com/api/v1/latlong/crimes/all/coords/{}/{}/distance/{}'.format(coordinates[1],coordinates[0],distance)
    res = requests.get(url)
    results = json.loads(res.content)
    return results
```

```python
dist = 1000
coord = (19.433735, -99.133334)
near_zocalo = crimes_loc(coord, dist)
near_zocalo
```

```python
near_zocalodf = pd.read_json(json.dumps(near_zocalo['rows']), orient='record')
near_zocalodf.head(5)
```

```python
def top_ranked_cuadrants(fecha):
    """
    Docstring
    description: Function that gives top crimes given a year
    input: year
    output: python dictionary with the information get
    """
    url = 'https://hoyodecrimen.com/api/v1/cuadrantes/crimes/all/top/counts?start_period1/{}-01/end_period1/{}-12'.format(fecha,fecha)
    res = requests.get(url)
    results = json.loads(res.content)
    return results
```

```python
fecha = (2016,2017,2018)
ranked2016 = top_ranked_cuadrants(fecha[0])
ranked2017 = top_ranked_cuadrants(fecha[1])
ranked2018 = top_ranked_cuadrants(fecha[2])
```

```python
rankfull = ranked2016['rows']+ranked2017['rows']+ranked2018['rows']
rankfulldf = pd.read_json(json.dumps(rankfull), orient='record')
```
